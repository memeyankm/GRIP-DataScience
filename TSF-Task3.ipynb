```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# importing the dataset
data = pd.read_csv('Iris.csv')
x = data.iloc[:, [1, 2, 3, 4]].values
```

```python
# finding optimum number of classifications
wcss = [] # Within Cluster Sum of Squares

for _ in range(1, 11):
    model_task3 = KMeans(n_clusters=_, init='k-means++', n_init=10, max_iter=300, random_state=0)
    model_task3.fit(x)
    wcss.append(model_task3.inertia_)

# plotting them into one graph
plt.figure()
plt.plot(range(1, 11), wcss, color='r')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method')
```

```python
# applying k-means to the dataset
model_task3 = KMeans(n_clusters = 3, init = 'k-means++', 
                     max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = model_task3.fit_predict(x)
```

```python
# visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'green', label = 'Iris-setosa')

plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')

plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], 
            s = 100, c = 'red', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(model_task3.cluster_centers_[:, 0], model_task3.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
